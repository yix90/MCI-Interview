2. The responses in study_101 is exactly the same as the responses in the sample database schema responses sheet. While I can technically do a copy and paste work, it is only right to process this exercise the raw way.

Here we go.

Actions to be taken:
1. Respondent_id for study_101 is clearly different from study_102 (can roughly tell from the sex option). Remove that column.
2. Number of children is a critical feature but study_101 does not have this particular data. We will include that but for the study_101 portion the data point for number of children will be indicated as a null value. Likewise for Q4 for study_102, it is missing.
3. We'll need to write a function to split the responses for Q3 into separate columns to better represent categorical data.
4. We need to standardize the date too. I shall make an assumption that the survey is done at the beginning of the month, and set it in the format of dd/mm/yyyy

```python
#Now lock it in
s1['date'] = s1['date'].apply(lambda x: pd.to_datetime(x, format='%Y-%d-%M').strftime('%d/%M/%Y'))
s1['study_id'] = 101
s2['date'] = s2['date'].apply(lambda x: pd.to_datetime(x, format='%M/%d/%Y').strftime('%d/%M/%Y'))
s2['study_id'] = 102
```

```python
#Missing values automatically get filled with NaN
combine = s1.append(s2)
combine
```

```python
for i in range(1,7):
    j = str(i)
    combine['Q3_'+j] = combine['Q3'].apply(lambda x: j in x).astype('int')
combine
#It works
```

```python
#Drop respondent id column, run once
combine.drop(['Q3'], axis=1, inplace=True)
combine.reset_index(drop=True, inplace=True)
```

```python
combine.rename(columns={'Number of children':'n_children'}, inplace=True)
```

```python
combine
```

This sheet is technically enough for everything but we shall input the data according to schema.

Create a new respondents table and a new responses table.

```python
#Number of children was not available in the schema, I'll put it in the respondents schema
respn1 = combine[['respondent_id','study_id','Sex','date','n_children']]
respn1.rename(columns={'respondent_id':'id', 'Sex':'sex'}, inplace=True)
```

```python
new_df = combine[['respondent_id','study_id','Q1','Q2','Q3_1','Q3_2','Q3_3','Q3_4','Q3_5','Q3_6','Q4','Q5']]
newer_df = pd.melt(new_df, id_vars=['respondent_id','study_id'],value_vars=['Q1','Q2','Q3_1','Q3_2','Q3_3','Q3_4','Q3_5','Q3_6','Q4','Q5'])
newer_df = newer_df.sort_values(['study_id','respondent_id'])
newer_df = newer_df.reset_index(drop=True)
newer_df = newer_df.reset_index() #Again for the column
newer_df['index'] = newer_df['index']+1
newer_df = newer_df.rename(columns={'index':'id'})
```

```python
import sqlite3
sqlite_db = './dbase.sqlite'
conn = sqlite3.connect(sqlite_db) 
```

```python
#Put the metadata schema inside first
# codebook.to_sql('codebook', con=conn, if_exists = 'replace', index = False)
#Sorry I failed to put in the codebook, it produced a formatting error.
question.drop(['Datatype'], axis=1, inplace=True) #That extra column wasn't part of it to begin with
question.to_sql('question', con=conn, if_exists = 'replace', index = False)
```

```python
newer_df = newer_df.rename(columns={'variable':'question_id','value':'response'})
```

```python
#Now the respondents and responses get added in!
respn1.to_sql('respondent', con=conn, if_exists = 'replace', index=False, dtype={'id':'INT','study_id':'INT','sex':'INT','n_children':'INT','date':'TEXT'})
newer_df.to_sql('response', con=conn, if_exists = 'replace', index=False, dtype={'id':'INT','study_id':'INT','respondent_id':'INT','question_id':'TEXT','response':'INT'})
```
